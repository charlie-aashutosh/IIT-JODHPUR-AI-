Creating dendrograms using Agglomerative Heirarchial Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/diabetes.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
import scipy.cluster.hierarchy as sch
plt.title("Dendrogram for Diabetes dataset")
dendrogram=sch.dendrogram(sch.linkage(data,method='ward'))
```

```python
new_data=pd.read_csv("/content/3.01. Country clusters.csv")
```

```python
new_data.head(6)
```

```python
df=new_data.iloc[:,1:3]
```

```python
df.head()
```

```python
plt.title("Dendrogram for Country cluster dataset")
dendrogram=sch.dendrogram(sch.linkage(df,method='ward')) #K=3
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(3) #here 3 indicates the number of clusters (K=3_)
```

```python
clusters=kmeans.fit_predict(df)
```
